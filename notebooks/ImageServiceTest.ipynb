# Graph Image Generator Service Example

## What's tihs?

This is an example how to use graph image generator service.  _The service code is still experimental!_

## Basic Setup

```python
import requests
from IPython.display import Image, SVG
from IPython.core.display import HTML
from IPython.display import (
    display_pretty, display_html, display_jpeg,
    display_png, display_json, display_latex, display_svg
)

# This is your service IP address.  You need to modify this!
BASE = 'http://192.168.99.100:5000/'
# BASE = 'http://52.35.133.2:5000/'

# Check server is actually working
res = requests.get(BASE)
print(res.json())

# Helper function to POST a local CX file
def post_file(url, file_name):
    with open(file_name, 'r') as f:
        return requests.post(url, data=f)
```

## Generate a PNG (bitmap) image from CX file

```python
bitmap_image_url = BASE + 'image'
res = post_file(bitmap_image_url, 'cx1.json')
Image(data=res.content)
```

## Generate SVG (vector) file from CX

```python
svg_image_url = BASE + 'image/svg'
res = post_file(svg_image_url, 'cx1.json')

# print(res.content)
SVG(data=res.content)
```

## Use Graphviz as renderer

```python
svg_image_url = BASE + 'image/graphviz/svg'
res = post_file(svg_image_url, 'cx1.json')
SVG(data=res.content)
```
